```python
import pyautogui
from PIL import Image
import cv2 as cv2
```

```python
def Capture_Event(event, x, y, flags, params):
	# If the left mouse button is pressed
	if event == cv2.EVENT_LBUTTONDOWN:
		# Print the coordinate of the 
		# clicked point
		print(f"({x}, {y})")
if __name__=="__main__":
	# Read the Image.
	img = cv2.imread('img.png', 1)
	# Show the Image
	cv2.imshow('image', img)
	# Set the Mouse Callback function, and call
	# the Capture_Event function.
	cv2.setMouseCallback('image', Capture_Event)
	# Press any key to exit
	cv2.waitKey(0)
	# Destroy all the windows
	cv2.destroyAllWindows()
```

```python
import requests
from bs4 import BeautifulSoup

page = requests.get("https://en.wikipedia.org/wiki/Subdivisions_of_Egypt")

src = page.content

soup = BeautifulSoup(src, 'lxml')
soup
```

```python
table = soup.find("table",{"class":"wikitable"}).find("tbody")
table
```

```python
Rows = table.find_all("tr")
Rows[1].contents[7].text
```

```python
Rows = table.find_all("tr")
header = ["City","population","Area","Density"]
population = []
for row in Rows[1:len(Rows)-1]:
    population.append({"City":row.contents[1].text.replace("\n","").strip(),
                       "population":(row.contents[5].text).replace("\n",""),
                       "Area":(row.contents[3].text).replace("\n",""),
                       "Density":((row.contents[7].text).replace("\n",""))})

print(population)
    
```

```python
import csv
with open("random.csv","w") as outputFile:
    dictWriter = csv.DictWriter(outputFile, header)
    dictWriter.writeheader()
    dictWriter.writerows(population)
```

```python
import csv

# Function to convert TSV to CSV with header
def tsv_to_csv_with_header(tsv_file, csv_file):
    with open(tsv_file, 'r', newline='') as tsvfile:
        with open(csv_file, 'w', newline='') as csvfile:
            tsvreader = csv.reader(tsvfile, delimiter='\t')
            csvwriter = csv.writer(csvfile)
            
            # Read the header row from TSV and write it to CSV
            header = next(tsvreader)
            csvwriter.writerow(header)
            
            # Write the remaining rows
            for row in tsvreader:
                csvwriter.writerow(row)

# Usage example
tsv_to_csv_with_header('locations2.tsv', 'location2.csv')
```
